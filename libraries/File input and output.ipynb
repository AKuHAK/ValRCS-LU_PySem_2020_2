```python
with open("../data/Veidenbaums.txt") as f1:
    for line in f1:
        print(line, end="")
```

```python
with open("../data/Veidenbaums.txt") as f1:
    data = f1.readlines()
```

```python
data[:10]
```

```python
with open("../data/Veidenbaums_clean.txt", "w") as f1:
    
    clean_data = [item for item in data if len(item.strip()) > 0]
```

```python
len(clean_data), len(data)
```

```python
clean_data[:10]
```

```python
with open("../data/Veidenbaums_clean.txt", "w") as f1:
    
    f1.writelines(clean_data)
```

```python
!cat ../data/Veidenbaums_clean.txt
```

```python
with open("../data/Veidenbaums.txt", "r") as f1, \
  open("../data/Veidenbaums_clean.txt", "w") as out_f:
        
        for line in f1:
            if len(line.strip()) > 0:
                out_f.write(line)
                
```
